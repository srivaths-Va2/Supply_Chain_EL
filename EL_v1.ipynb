```python
#distribution of min price, max price and modal price through pie chart
from seaborn._core.properties import FontSize
labels='Min Price', 'Max Price', 'Modal Price'
size =[215,130, 245]
colors = ['gold', 'yellowgreen', 'lightcoral']
textprops={"fontsize":18}
explode= (0.1, 0, 0)
#plot
plt.pie(size, explode= explode, labels=labels, colors=colors, shadow= True, autopct='%1.1f%%', textprops=textprops )
plt.axis('equal')
plt.show()
```

```python
#LET'S FIND OUTLIERS USING BOX PLOT
sns.set (font_scale=1)
price= df[['Min Price', 'Max Price', 'Modal Price']]
sns.boxplot (data=df)
plt.xlabel('Price Type')
plt.ylabel('Price')
plt.title('Price Distribution')
plt.show()
```

```python
#hist plot of modal price and vs. frequency
plt.figure (figsize=(10,6))
sns.histplot(data=df, x='Modal Price', bins=25, color='skyblue')
plt.xlabel('Modal Price')
plt.ylabel('frequency')
plt.title('Modal price distribution')
plt.show()
```

```python
#scatter plot of max price and min price
plt.figure (figsize=(10,6))
sns.scatterplot(data=df, x='Max Price', y='Min Price', color='blue')
plt.xlabel('Max Price')
plt.ylabel('Min Price')
plt.title('Max Price and Min Price distribution')
plt.show()
```

```python
top_10_highest = df.nlargest(10,'Modal Price')
print(top_10_highest)
```

```python
#using bar plot
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
sns.barplot(data= top_10_highest, x='State', y='Modal Price')
plt.xlabel('State')
plt.ylabel('Modal Price')
plt.title(' State- wise Top 10 highest Modal price')
plt.show()
```

```python
#using bar plot
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.barplot(data= top_10_highest, x='Market', y='Modal Price')
plt.xlabel('Market')
plt.ylabel('Modal Price')
plt.title('Top 10 highest Modal price by market')
plt.xticks(rotation=90)
plt.show()
```

```python
#using bar plot
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.barplot(data= top_10_highest, x='Commodity', y='Modal Price')
plt.xlabel('Commodity')
plt.ylabel('Modal Price')
plt.title('Top 10 highest Modal price by Commodity')
plt.xticks(rotation=90)
plt.show()
```

```python
#using bar plot
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.barplot(data= top_10_highest, x='Variety', y='Modal Price')
plt.xlabel('Variety')
plt.ylabel('Modal Price')
plt.title('Top 10 highest Modal price by Variety')
plt.xticks(rotation=90)
plt.show()
```

```python
#using violin plot
#it help us to see both the distribution of data in terms of kernel density estimation and box plot)
plt.figure(figsize=(6,4))
sns.violinplot(data=df, y='Modal Price', palette= 'rainbow')
plt.ylabel('Modal Price')
plt.title('Modal Price Ditribution')
plt.tight_layout()
plt.show()
```

```python
price_columns= ['Min Price', 'Max Price', 'Modal Price']
plt.figure(figsize=(6,4))
sns.pairplot(data=df, vars= price_columns)
plt.tight_layout()
plt.show()
```

```python
price_columns=['Min Price', 'Max Price', 'Modal Price']
price_correlation= df[price_columns].corr()
plt.figure(figsize=(8,6))
sns.heatmap(price_correlation, annot= True, cmap= 'coolwarm', square= True)
plt.title('Price Correlation')
plt.tight_layout()
plt.show()
```
